```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/rep_summary.csv')
```

```python
df['Party'].fillna("Unknown", inplace=True)
```

```python
counter = Counter(df['Party'])
counter['Democratic Party'] += counter['Democratic']
counter['Republican Party'] += counter['Republican']
counter['Unknown'] += counter['nan']
counter.pop('Democratic', None)
counter.pop('Republican', None)
counter.pop('nan', None)
parties = counter.keys()
party_counts = counter.values()
print(counter)
# Plot histogram using matplotlib bar().
indexes = np.arange(len(parties))
width = .8
plt.bar(indexes, party_counts, width, color=[(0, 0, 1, 0.6), (0.5, 1, 0.6, 0.6), (0.5, 0.5, 0.6, 0.6), (1, .1, .1, .6)])
plt.xticks(indexes, parties)
plt.title("Number of Politicians by Party")
plt.savefig('politicalParties.png')
plt.show()
print(indexes)
```

```python
counts= df.count()
media = ['Twitter', 'Facebbook', 'Website']
counts = [counts['Twitter'], counts['Facebook'],counts['Website']]

indexes = np.arange(len(media))
width = .8
plt.bar(indexes, counts, width, color=[(0, 0.674, 0.929,1), (0.235, 0.352, 0.6,1), (1, .2, 0.2, 0.6)])
plt.xticks(indexes, media)
plt.title("Number of Politicians with Online Presence")
plt.savefig('socialmedias.png')
plt.show()
```
